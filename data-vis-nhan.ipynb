```python
## Extract the text file and fill values

time_values = []
accx_values = []
accy_values = []
accz_values = []

with open('data/20241226115026.txt', 'r') as file:
    # initialization of t0
    h_0 = -99.99
    m_0 = -99.99
    s_0 = -99.99
    prev_second = -99
    # counter
    counter=0
    
    for line in file:
        columns = line.split()
        columns_new = []
        if counter > 0: # re-combine the timestamp 
            columns_new = [columns[0]+" "+columns[1]] + columns[2:]
        else: 
            columns_new = columns
        
        if counter >= 1: 
            
            date = columns_new[0].split()[0]
            time = columns_new[0].split()[1]

            hours, minutes, seconds = map(float, time.split(':'))

            if counter == 1: prev_second = int(seconds)
            
            if counter == 2: # just in case the first time stamp gets messed up
                h_0 = hours
                m_0 = minutes
                s_0 = seconds

            accx_values.append(float(columns_new[2]))
            accy_values.append(float(columns_new[3]))
            accz_values.append(float(columns_new[4]))

            time_in_seconds = (hours-h_0)*3600. + (minutes-m_0)*60. + (seconds-s_0)
            

            if int(seconds) != int(prev_second):
                print("new second!")
            
            time_values.append(time_in_seconds)
            
            print(date, hours, minutes, seconds, time_in_seconds, int(seconds), int(prev_second))
            prev_second = int(seconds)
            
        counter += 1
        
        # if counter > 10: break # for debugging
```
